# Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('adult.csv')
```

# Exploration

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.head(10)
```

# Preprocessing

## Missing Values

```python
df = df.replace('?', np.nan)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy = 'most_frequent')
```

```python
df = pd.DataFrame(imputer.fit_transform(df), columns = df.columns)
```

## Encoding

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
hot_encoder = OneHotEncoder(sparse_output = False, handle_unknown = 'ignore', drop = 'first').set_output(transform = 'pandas')
```

```python
df = pd.concat([df.drop('gender', axis = 1).reset_index(drop = True), hot_encoder.fit_transform(df[['gender']]).astype(int)], axis = 1)
```

```python
categorical_columns = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'native-country', 'income']

for column in categorical_columns:
  df[column] = label_encoder.fit_transform(df[column])
```

## Data Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('income', axis = 1)
y = df['income']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123)
```

## Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns = X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns = X_test.columns)
```

# Training

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(metric = 'euclidean')
```

```python
model.fit(X_train, y_train)
```

# Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report
```

## Predictions

```python
predictions = model.predict(X_test)
```

```python
predictions
```

## Accuracy

```python
accuracy = accuracy_score(y_test, predictions)
```

```python
print('Model Accuracy: ', round(accuracy, 3) * 100, '%\n')
print(classification_report(y_test, predictions))
```
